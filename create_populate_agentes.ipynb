```python
--- CREATE TABLE agentes --
drop table IF EXISTS e2014.agentes;
CREATE TABLE e2014.agentes(
  id integer NOT NULL DEFAULT nextval('id'),
	cod_agente	varchar,
	tipo		varchar,
	cnpj		varchar,
	sigla_uf	varchar,
	sigla_ue	varchar,
	numero_partido	varchar,
	sigla_partido	varchar,
	nome_agente	varchar,
	cod_candidatura	varchar,
	nome_candidatura	varchar,
	cargo           varchar,
	apto            varchar,
	codigo_cargo    varchar,
	grupo           varchar,
	nome_grupo      varchar
);

CREATE  INDEX cnpj_idx ON e2014.agentes (cnpj);

CREATE  INDEX nome_agente_idx ON e2014.agentes (nome_agente);
```

```python
--------------------------------------------
-- AJUSTA NUMERO_PARTIDO EM CF
with partidos as
(
  select
   numero,
   sigla
  from 
    e2014.partidos
)
UPDATE e2014.agentes as a
set 
    numero_partido = p.numero
from 
    partidos as p
where
    tipo = 'CF'
    and
    p.sigla = a.sigla_partido 
;

-- AJUSTA NUMERO_PARTIDO EM DE
with partidos as
(
  select
   numero,
   sigla
  from 
    e2014.partidos
)
UPDATE e2014.agentes as a
set 
    numero_partido = p.numero
from 
    partidos as p
where
    tipo = 'DE'
    and
    p.sigla = a.sigla_partido 
;
```

```python
-- AJUSTA SIGLA_PARTIDO EM CF
with partidos as
(
  select
   numero,
   sigla
  from 
   e2014.partidos
)
UPDATE e2014.agentes as a
set 
    sigla_partido = p.sigla
from 
    partidos as p
where
    tipo in ('CF', 'CC')
    and
    sigla_partido = ''
    and
    p.numero = a.numero_partido 
;


-- AJUSTA SIGLA_PARTIDO EM DE
with partidos as
(
  select
   numero,
   sigla
  from 
   e2014.partidos 
)
UPDATE e2014.agentes as a
set 
    sigla_partido = p.sigla
from 
    partidos as p
where
    tipo = 'DE'
    and
    sigla_partido = ''
    and
    p.numero = a.numero_partido 
;
```

```python
-- ASSOCIA DADOS CANDIDATURA A CC
with candidaturas as
(
  select 
    cnpj_candidatura,
    sigla_uf,
    sigla_ue,
    numero_partido,
    sigla_partido,
    cod_candidatura,
    tipo_candidatura,
    nome_candidato,
    nome_candidatura,
    descricao_cargo,
    codigo_cargo,
    apto
  from
    e2014.candidaturas as c
) 
UPDATE e2014.agentes as a
  set 
    cod_agente = c.cod_candidatura,
    sigla_uf = c.sigla_uf,
    sigla_ue = c.sigla_ue,
    numero_partido = c.numero_partido,
    sigla_partido  = c.sigla_partido,
    cod_candidatura = c.cod_candidatura,
    nome_candidatura = c.nome_candidatura,
    cargo = c.descricao_cargo,
    codigo_cargo = c.codigo_cargo,
    apto         = c.apto,
    grupo        = c.tipo_candidatura,
    nome_grupo   = c.nome_candidatura
from
  candidaturas as c
where
    a.tipo = 'CC'
    and
    c.cnpj_candidatura like a.cnpj
;
```

```python
/*--
FAZER BUSCA PELO NOME PARA OS 'CC' QUE FICARAM SEM COD-CANDIDATURA
PELO NOME
--*/ 


with candidaturas as
(
  select 
    c.cnpj_candidatura,
    c.sigla_uf,
    c.sigla_ue,
    c.numero_partido,
    c.sigla_partido,
    c.cod_candidatura,
    c.tipo_candidatura,
    c.nome_candidato,
    c.nome_candidatura,
    c.descricao_cargo,
    c.codigo_cargo
  from
    e2014.candidaturas as c
)   
update e2014.agentes as a
set 
    cod_agente = c.cod_candidatura,
    sigla_uf = c.sigla_uf,
    sigla_ue = c.sigla_ue,
    numero_partido = c.numero_partido,
    sigla_partido  = c.sigla_partido,
    cod_candidatura = c.cod_candidatura,
    nome_candidatura = c.nome_candidatura,
    cargo = c.descricao_cargo,
    codigo_cargo = c.codigo_cargo,
    grupo        = c.tipo_candidatura,
    nome_grupo   = c.nome_candidatura,
    nome_agente  = c.nome_candidatura
from candidaturas as c
where
    a.nome_agente like '%'||trim(c.nome_candidato)||'%'
    --is_nome_igual(c.nome_candidato,a.nome_agente)
; 



-- ajusta cod_agente e nome agente, grupo e nome_grupo se CF 
update e2014.agentes as a
set cod_agente = 'AP'||sigla_ue||numero_partido,
    nome_agente = 'Comitê Financeiro ('||sigla_partido||'/'||sigla_ue||')',
    grupo = 'AP',
    nome_grupo = 'Agente Partidário ('||sigla_partido||'/'||sigla_ue||')'
where tipo  = 'CF'
;
  
-- ajusta cod_agente e nome agente, grupo e nome_grupo se DE
update e2014.agentes as a
set cod_agente = 'AP'||sigla_ue||numero_partido,
    nome_agente = 'Direção Partidária ('||sigla_partido||'/'||sigla_ue||')',
    grupo = 'AP',
    nome_grupo = 'Agente Partidário ('||sigla_partido||'/'||sigla_ue||')'
where tipo = 'DE'
;
```

```python
----------------   IMPRTANTE -----------------------------------------------------------------------------
-- ACERTAR SIGLA UF PARA OS CC cod_agente = '' A PARTIR DE SIGLA UE --- REVISAR A GERACAO DE SIGLA US EM TODAS AS TABELAS
-----------------------------------------------------------------------------------------------------------
with municipios as
( select sigla_uf, sigla_ue, nome_ue from e2014.municipios )
update e2014.agentes as a
set sigla_uf = m.sigla_uf
from municipios as m
where m.sigla_ue = a.sigla_ue
;

-- estabelece cod_agente e nome agente, grupo e nome_grupo se CC 
with agentes as (
select cnpj, grupo,sigla_uf,sigla_partido,numero_partido from e2014.agentes
) 
update e2014.agentes as aj
set 
    cod_agente = 'AP'||a.sigla_uf||a.numero_partido,
    nome_agente = 'Comitê Financeiro ('||a.sigla_partido||'/'||a.sigla_uf||')',
    grupo = 'AP',
    nome_grupo =  'Agente Partidário ('||a.sigla_partido||'/'||a.sigla_uf||')'
from agentes as a
where aj.tipo = 'CC' 
and aj.grupo = ''
and aj.sigla_uf <> '#NULO'
and aj.cnpj = '%'||a.cnpj||'%'
;
```
